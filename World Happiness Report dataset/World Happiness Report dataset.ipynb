```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
```

```python
# Step 2: Load the Data
# Load the dataset
url = 'https://raw.githubusercontent.com/PhilippeCodes/World-Happiness-Report-Data-Analysis/master/World%20Happiness%20Report.csv'
df = pd.read_csv(url)
```

```python
# Step 3: Understand the Data
# Display the first few rows
print(df.head())

# Get basic information
print(df.info())

# Summary statistics
print(df.describe(include='all'))
```

```python
# Step 4: Handle Missing Data
# Check for missing values
print(df.isnull().sum())

# Handle missing values
# For simplicity, you might drop rows with missing values
df.dropna(inplace=True)  # Alternatively, you can use df.fillna() for imputation
```

```python
# Step 5: Data Visualization
# Distribution of Happiness Scores
sns.histplot(df['Happiness Score'], kde=True, bins=30)
plt.title('Distribution of Happiness Scores')
plt.show()

# Happiness by Region
plt.figure(figsize=(12, 8))
sns.boxplot(x='Region', y='Happiness Score', data=df)
plt.title('Happiness Scores by Region')
plt.xticks(rotation=90)
plt.show()

# Correlation between variables
plt.figure(figsize=(12, 8))
sns.heatmap(df[['Happiness Score', 'GDP per Capita', 'Social Support', 'Health Life Expectancy', 'Freedom', 'Generosity', 'Corruption']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Step 6: Univariate Analysis
# Distribution of GDP per Capita
sns.histplot(df['GDP per Capita'], kde=True, bins=30)
plt.title('Distribution of GDP per Capita')
plt.show()

# Distribution of Social Support
sns.histplot(df['Social Support'], kde=True, bins=30)
plt.title('Distribution of Social Support')
plt.show()
```

```python
# Step 7: Bivariate Analysis
# Happiness vs GDP per Capita
sns.scatterplot(x='GDP per Capita', y='Happiness Score', data=df)
plt.title('Happiness Score vs GDP per Capita')
plt.show()

# Happiness vs Social Support
sns.scatterplot(x='Social Support', y='Happiness Score', data=df)
plt.title('Happiness Score vs Social Support')
plt.show()
```

```python
# Step 8: Multivariate Analysis
# Pair plot for selected features
sns.pairplot(df[['Happiness Score', 'GDP per Capita', 'Social Support', 'Health Life Expectancy', 'Freedom']])
plt.show()
```

```python
# Step 9: Identify and Handle Outliers
# Box plot to identify outliers in Happiness Score
sns.boxplot(x=df['Happiness Score'])
plt.title('Boxplot of Happiness Scores')
plt.show()

# Removing outliers from Happiness Score
Q1 = df['Happiness Score'].quantile(0.25)
Q3 = df['Happiness Score'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['Happiness Score'] < (Q1 - 1.5 * IQR)) | (df['Happiness Score'] > (Q3 + 1.5 * IQR)))]

# Box plot to identify outliers in GDP per Capita
sns.boxplot(x=df['GDP per Capita'])
plt.title('Boxplot of GDP per Capita')
plt.show()
```

```python
# Step 10: Feature Engineering
# Create additional features or normalize data if needed
# For example, normalize the GDP per Capita
df['GDP per Capita Normalized'] = (df['GDP per Capita'] - df['GDP per Capita'].mean()) / df['GDP per Capita'].std()
```

```python
# Step 11: Summary and Insights
# Summarize key findings
print("Key Insights:")

# Happiness Score distribution
happiness_dist = df['Happiness Score'].describe()
print(f"Happiness Score Distribution:\n{happiness_dist}")

# Happiness by Region
happiness_by_region = df.groupby('Region')['Happiness Score'].mean()
print(f"Average Happiness Score by Region:\n{happiness_by_region}")

# Correlations
correlations = df[['Happiness Score', 'GDP per Capita', 'Social Support', 'Health Life Expectancy', 'Freedom', 'Generosity', 'Corruption']].corr()
print(f"Correlation Matrix:\n{correlations}")

# Insights from scatter plots
print("Happiness Score appears positively correlated with GDP per Capita and Social Support.")
```

Findings:
1. Happiness Scores: The dataset shows the distribution and average happiness scores across different regions.
2. Regional Differences: Certain regions have significantly different average happiness scores.
3. Correlations: Happiness is often positively correlated with factors like GDP per Capita and Social Support.
4. Temporal and Feature Trends: Analyzing the correlations and feature distributions provides insights into what factors contribute most to happiness.
